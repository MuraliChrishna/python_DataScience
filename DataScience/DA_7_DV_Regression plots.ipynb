```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
from scipy import stats
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip',tips)
# that is a liner regression line
# also know as scatter plot
```

```python
sns.lmplot('total_bill','tip',tips,order=4,
           scatter_kws={'marker':'*','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
# if we dont want linear line we can indicate the order
```

```python
sns.lmplot('total_bill','tip',tips, fit_reg=False)
```

```python
tips.head()
```

```python
tips['tips_pect']=100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('size','tips_pect',tips)
```

```python
# we can add a gitter to make the lmplot more readable
```

```python
url ='http://en.wikipedia.org/wiki/jitter'
```

```python
sns.lmplot('size','tips_pect',tips,x_jitter=.1)
```

```python
# we can also extimate the estimate the tendency of the each bin 
```

```python
sns.lmplot('size','tips_pect',tips,x_estimator= np.mean)
```

```python
sns.lmplot('total_bill','tips_pect', tips, hue= 'sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','tips_pect',tips,hue='day')
```

```python
# sea born also supports l;ocal regression marhker fitting
```

```python
# wiki for local regression 
```

```python
sns.lmplot('total_bill','tips_pect',tips,lowess=True, line_kws ={'color':'black'})
```

```python
#lm plot uses regressionplots

sns.regplot('total_bill','tips_pect',tips)
```

```python
fig,(axis1, axis2) = plt.subplots(1,2,sharey =True)
sns.regplot('total_bill','tips_pect', tips, ax = axis1)
sns.violinplot(tips['tips_pect'],tips['size'],color = 'red',ax = axis2)
```
