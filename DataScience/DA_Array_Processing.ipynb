```python
# numpy where 

a = np.array([1,2,3,4])
```

```python
b = np.array([100,200,300,400])
```

```python
condition = np.array([True,True,False,False])
```

```python
answer = [(a_val if cond else b_val) for a_val,b_val, cond in zip(a,b,condition)]
# chose a value from a if condition is True 
# choose b value from b if the condition is False
```

```python
answer
```

```python
answer2 = np.where(condition,a,b)
# this is the short form of above command
```

```python
answer2
```
